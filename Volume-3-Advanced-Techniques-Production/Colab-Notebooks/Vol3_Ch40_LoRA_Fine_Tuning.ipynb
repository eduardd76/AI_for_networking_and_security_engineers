Low-Rank Adaptation math and parameter efficiency.


```python
import numpy as np
print("LoRA ready")
```

```python
np.random.seed(42)
W=np.random.randn(64,64)
for r in [2,4,8,16]:
    U,S,Vt=np.linalg.svd(W,full_matrices=False)
    A=U[:,:r]*np.sqrt(S[:r]); B=np.sqrt(S[:r])[:,None]*Vt[:r,:]
    err=np.linalg.norm(W-A@B,"fro")/np.linalg.norm(W,"fro")
    saving=(1-(A.size+B.size)/W.size)*100
    print(f"rank={r:2d}: error={err:.4f}, saving={saving:.1f}%")
```

```python
d,r,L=4096,8,32
orig=d*d*L*2; lora_p=2*d*r*L*2
print(f"Original: {orig:,} | LoRA: {lora_p:,} | {lora_p/orig*100:.2f}%")
np.random.seed(7)
A=np.random.randn(16,r)*0.01; B=np.zeros((r,16)); T=np.random.randn(16,16)
loss0=None
for _ in range(30):
    pred=A@B; loss=np.mean((pred-T)**2)
    if loss0 is None: loss0=loss
    A-=0.01*2*(pred-T)@B.T/pred.size; B-=0.01*2*A.T@(pred-T)/pred.size
print(f"Training: {loss0:.4f} -> {loss:.4f} ({(1-loss/loss0)*100:.1f}% reduction)")
```
