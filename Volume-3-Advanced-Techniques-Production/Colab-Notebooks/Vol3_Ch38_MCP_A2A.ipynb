# Chapter 38: Agent Communication Protocols (MCP and A2A)
MCP tool definitions and A2A message passing.


```python
import json, uuid
from datetime import datetime
print("MCP/A2A ready")
```

```python
def make_tool(name,desc,params):
    return {"name":name,"description":desc,"inputSchema":{"type":"object","properties":{p:{"type":"string"} for p in params},"required":params}}
tools=[make_tool("query_device","Query device",["hostname","metric"]),make_tool("apply_policy","Apply policy",["policy_id","target"])]
print(json.dumps(tools,indent=2))
```

```python
def rpc(method,params):
    return {"jsonrpc":"2.0","id":str(uuid.uuid4())[:8],"method":method,"params":params}
req=rpc("tools/call",{"name":"query_device","arguments":{"hostname":"sw1","metric":"cpu"}})
print(json.dumps(req,indent=2))
msg={"id":str(uuid.uuid4())[:8],"ts":datetime.utcnow().isoformat(),"sender":"orchestrator","receiver":"net-agent","type":"TASK_ASSIGN","payload":{"task":"audit_acls"}}
print(json.dumps(msg,indent=2))
```
