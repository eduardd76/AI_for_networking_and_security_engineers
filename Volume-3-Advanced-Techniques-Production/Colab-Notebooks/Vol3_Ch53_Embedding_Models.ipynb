# Chapter 53: Embedding Models and Semantic Similarity
Sentence encoding, cosine similarity, and clustering.


```python
import numpy as np
from collections import defaultdict
print("Embedding Models ready")
```

```python
def encode(sents,dim=64):
    embs=[]
    for s in sents:
        np.random.seed(abs(hash(s))%2**31)
        v=np.random.randn(dim); v/=np.linalg.norm(v); embs.append(v)
    return np.array(embs)
sents=["Firewall blocked traffic","Firewall denied connection","BGP route flapping","BGP session down","CPU threshold exceeded"]
embs=encode(sents)
print("Embedding shape:",embs.shape)
```

```python
sim=embs@embs.T
np.fill_diagonal(sim,0)
i,j=divmod(sim.argmax(),len(sents))
print("Most similar (score="+str(round(float(sim[i,j]),3))+"):")
print("  ",sents[i])
print("  ",sents[j])
labels=[0 if "Firewall" in s or "BGP" in s else 1 for s in sents]
clusters=defaultdict(list)
for idx,l in enumerate(labels): clusters[l].append(sents[idx][:45])
print("Clusters:")
for cid,members in clusters.items(): print("  Cluster",cid,":",members)
```
