BLEU, faithfulness, and consistency metrics.


```python
import numpy as np
from collections import Counter
print("LLM Evaluation ready")
```

```python
def bleu2(ref,hyp):
    r=ref.lower().split(); h=hyp.lower().split()
    if len(h)<2: return 0.0
    rg=Counter(tuple(r[i:i+2]) for i in range(len(r)-1))
    hg=Counter(tuple(h[i:i+2]) for i in range(len(h)-1))
    inter=hg&rg; m=sum(inter.values()); t=sum(hg.values())
    return m/t if t else 0.0
ref="The firewall blocks unauthorized traffic"
hyp="The firewall denies unauthorized traffic"
print(f"BLEU-2: {bleu2(ref,hyp):.3f}")
```

```python
def embed(text,dim=64):
    np.random.seed(abs(hash(text))%2**31)
    v=np.random.rand(dim); return v/np.linalg.norm(v)
def cos(a,b): return float(np.dot(a,b))
ctx="BGP route reflectors reduce mesh complexity."
for r in ["Route reflectors simplify BGP mesh.","VLANs segment broadcast domains."]:
    print(f"Faithfulness {cos(embed(ctx),embed(r)):.3f}: {r}")
ans=["VLAN is a logical network segment.","VLANs create isolated broadcast domains.","Virtual LAN isolates traffic."]
embs=[embed(a) for a in ans]
s=[float(np.dot(embs[i],embs[j])) for i in range(len(embs)) for j in range(i+1,len(embs))]
print(f"Consistency: {np.mean(s):.3f}")
```
