# Chapter 42: Tool Use and Function Calling
JSON Schema tools, routing, and structured output parsing.


```python
import json
print("Tool Use ready")
```

```python
TOOLS={
    "ping_host":{"desc":"Ping host","required":["host"]},
    "get_iface_stats":{"desc":"Interface stats","required":["device","iface"]},
    "lookup_cve":{"desc":"CVE lookup","required":["cve_id"]},
}
def route(name,args):
    if name not in TOOLS: return {"error":f"Unknown: {name}"}
    miss=[r for r in TOOLS[name]["required"] if r not in args]
    if miss: return {"error":f"Missing: {miss}"}
    if name=="ping_host": return {"latency_ms":12.4,"host":args["host"]}
    if name=="get_iface_stats": return {"rx_bytes":1024000,"tx_bytes":512000}
    if name=="lookup_cve": return {"cvss":8.1,"patch_available":True}
for n,a in [("ping_host",{"host":"8.8.8.8"}),("get_iface_stats",{"device":"sw1","iface":"Gi0/1"}),("lookup_cve",{"cve_id":"CVE-2024-1234"})]:
    print(f"{n}: {route(n,a)}")
```

```python
def parse_out(raw):
    try: d=json.loads(raw); return {"status":"ok","fields":list(d.keys())}
    except Exception as e: return {"status":"error","msg":str(e)}
for raw in [json.dumps({"action":"block","target":"1.2.3.4"}),json.dumps({"alert_id":"A-1042"}),"not-json"]:
    r=parse_out(raw); print(f"[{r[chr(115)+chr(116)+chr(97)+chr(116)+chr(117)+chr(115)]}] {raw[:30]}")
```
