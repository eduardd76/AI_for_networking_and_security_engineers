# Chapter 37: GraphRAG
Graph-Augmented Retrieval for richer AI context.


```python
import networkx as nx
import numpy as np
print("GraphRAG ready")
```

```python
G=nx.DiGraph()
nodes=[("Router-A","Device"),("Subnet-10","Network"),("FW1","Device")]
for n,t in nodes: G.add_node(n,ntype=t)
G.add_edge("Router-A","Subnet-10",rel="connects")
G.add_edge("FW1","Subnet-10",rel="protects")
print("Nodes:",list(G.nodes()))
print("Edges:",list(G.edges()))
```

```python
def rag_ctx(e,G,hops=2):
    seen=set(); front={e}
    for _ in range(hops):
        nxt=set()
        for n in front: nxt.update(G.successors(n)); nxt.update(G.predecessors(n))
        seen.update(nxt); front=nxt
    return list(seen)
print("Context:",rag_ctx("Router-A",G))
a=set(G.neighbors("Router-A")); b=set(G.neighbors("FW1"))
u=a|b; j=len(a&b)/len(u) if u else 0
print(f"Jaccard: {j:.3f}")
```
