# Chapter 26: Security Data Analysis
Statistical analysis of security events, correlation matrices, risk scoring.


```python
import random, math
random.seed(42)
print("Security Analysis ready")
```

```python
events=[{"type":random.choice(["auth_fail","port_scan","malware","data_exfil","policy"]),"severity":random.randint(1,10),"count":random.randint(1,50)} for _ in range(100)]
by_type={}
for e in events:
    t=e["type"]
    if t not in by_type: by_type[t]={"count":0,"severity_sum":0}
    by_type[t]["count"]+=e["count"]
    by_type[t]["severity_sum"]+=e["severity"]
print("Security event summary:")
for t,v in sorted(by_type.items()): print(" "+t+": count="+str(v["count"])+" avg_sev="+str(round(v["severity_sum"]/max(v["count"],1),1)))
```

```python
def risk_index(events):
    high=sum(e["count"] for e in events if e["severity"]>=8)
    total=sum(e["count"] for e in events)
    diversity=len(set(e["type"] for e in events))
    return round(math.log1p(high)*diversity/max(total,1)*100,2)
ri=risk_index(events)
print("Risk Index:",ri)
print("Risk Level:","HIGH" if ri>5 else "MEDIUM" if ri>2 else "LOW")
```
