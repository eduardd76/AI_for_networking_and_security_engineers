# Chapter 24: AI-Powered Log Analysis
Pattern detection, anomaly identification, automated log correlation.


```python
import random
random.seed(42)
print("AI Log Analysis ready")
```

```python
logs=[{"ts":i,"user":"u"+str(random.randint(1,5)),"action":random.choice(["login","file_read","file_write","delete","admin"]),"ip":"192.168."+str(random.randint(1,5))+"."+str(random.randint(1,100))} for i in range(200)]
user_actions={}
for l in logs:
    k=l["user"]
    if k not in user_actions: user_actions[k]=[]
    user_actions[k].append(l["action"])
print("Users:",len(user_actions))
for u,actions in sorted(user_actions.items()): print(" "+u+": "+str(len(actions))+" actions, admin="+str(actions.count("admin")))
```

```python
def anomaly_score(user,actions):
    score=0
    if actions.count("admin")>5: score+=3
    if actions.count("delete")>10: score+=2
    if len(set(actions))==1 and len(actions)>20: score+=2
    return score
for u,a in user_actions.items():
    s=anomaly_score(u,a)
    if s>0: print("ANOMALY "+u+" score="+str(s))
```
