# Building an Agent from Scratch
This notebook demonstrates how to build a simple, consistent agent capable of problem-solving and learning from tasks.

```python
# Define a basic agent that learns from task execution
class SimpleAgent:
    def __init__(self):
        self.knowledge_base = []
    
    def make_decision(self, task):
        print(f'Decision-making for task: {task}')
        # Simplified decision-making
        return 'Executed task' if 'network' in task else 'Task not recognized'
    
    def execute_task(self, task):
        decision = self.make_decision(task)
        print(decision)
        if decision == 'Executed task':
            self.knowledge_base.append(task)
        return decision
    
    def review_performance(self):
        print('Reviewing Performance')
        for t in self.knowledge_base:
            print(f'Learned task: {t}')

# Instantiate the agent
agent = SimpleAgent()
# Execute a network-related task
agent.execute_task('Analyze network traffic')
# Review the agent's performance
agent.review_performance()
```

## Key Concepts
- **Decision Making**: The agent evaluates tasks and makes logical decisions based on task content.
- **Knowledge Base**: Gains experience by storing completed tasks.
- **Performance Review**: Regularly reviews completed tasks to identify and consolidate learning.
