# Chapter 18: Hybrid Retrieval Systems
Combining dense and sparse retrieval, reciprocal rank fusion.


```python
import random
random.seed(42)
print("Hybrid Retrieval ready")
```

```python
corpus=[{"id":i,"text":"article about "+random.choice(["ML","networking","security","cloud"])} for i in range(50)]
def sparse_search(q,docs,k=10):
    terms=q.lower().split()
    scored=[(d,sum(1 for t in terms if t in d["text"].lower())) for d in docs]
    return [d for d,s in sorted(scored,key=lambda x:-x[1])[:k]]
def dense_search(q,docs,k=10):
    return sorted(docs,key=lambda x:random.random())[:k]
sparse=sparse_search("ML security",corpus)
dense=dense_search("ML security",corpus)
print("Sparse:",len(sparse),"Dense:",len(dense))
```

```python
def rrf(lists,k=60):
    scores={}
    for lst in lists:
        for rank,doc in enumerate(lst,1):
            did=doc["id"]
            scores[did]=scores.get(did,0)+1/(k+rank)
    return sorted(scores.items(),key=lambda x:-x[1])
fused=rrf([sparse,dense])
print("Fused top 5:")
for did,score in fused[:5]: print(" doc="+str(did)+" rrf="+str(round(score,4)))
```
