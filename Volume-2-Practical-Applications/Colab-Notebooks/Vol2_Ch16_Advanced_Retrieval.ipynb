# Chapter 16: Advanced Retrieval Techniques
Hybrid search, re-ranking, contextual compression for RAG pipelines.


```python
import random
random.seed(42)
print("Advanced Retrieval ready")
```

```python
docs=[{"id":i,"text":"Document "+str(i)+" about "+random.choice(["networking","security","AI","cloud","DevOps"]),"score":round(random.uniform(0.1,0.99),2)} for i in range(20)]
query="AI security"
bm25_scores=sorted(docs,key=lambda x:-x["score"])[:10]
print("BM25 top results:",len(bm25_scores))
for d in bm25_scores[:3]: print(" doc="+str(d["id"])+" score="+str(d["score"]))
```

```python
def rerank(docs, boost_terms):
    for d in docs:
        boost = sum(1 for t in boost_terms if t in d["text"].lower())
        d["final"] = d["score"] + boost * 0.1
    return sorted(docs, key=lambda x: -x["final"])
reranked = rerank(bm25_scores, ["AI","security"])
print("Reranked top 3:")
for d in reranked[:3]: print(" doc="+str(d["id"])+" final="+str(round(d["final"],2)))
```
