# Chapter 58: Network Traffic Analysis and Anomaly Detection
NetFlow features, statistical baseline, Z-score anomaly scoring.


```python
import numpy as np
import random
print("Traffic Analysis ready")
```

```python
random.seed(42)
n=1000
normal_bytes=[random.randint(100,5000) for _ in range(int(n*0.95))]
anomaly_bytes=[random.randint(50000,500000) for _ in range(int(n*0.05))]
all_bytes=normal_bytes+anomaly_bytes
mean_b=np.mean(all_bytes); std_b=np.std(all_bytes)
print("Flow stats: mean="+str(round(mean_b,0))+" std="+str(round(std_b,0)))
zscores=[(b,abs(b-mean_b)/(std_b+1e-9)) for b in all_bytes]
anomalies=[(b,z) for b,z in zscores if z>3]
print("Anomalies detected:",len(anomalies),"of",n)
print("Anomaly rate:",round(len(anomalies)/n*100,2),"%")
for b,z in sorted(anomalies,key=lambda x:-x[1])[:3]: print("  bytes="+str(b)+" zscore="+str(round(z,2)))
```

```python
port_counts={}
for _ in range(n): p=random.choice([80,80,80,443,443,22,53,4444,31337]); port_counts[p]=port_counts.get(p,0)+1
print("Top ports:")
for port,cnt in sorted(port_counts.items(),key=lambda x:-x[1])[:5]: print("  port="+str(port)+" count="+str(cnt))
suspicious_ports={4444,31337,8888,9999}
for port,cnt in port_counts.items():
    if port in suspicious_ports: print("  ALERT: Suspicious port "+str(port)+" seen "+str(cnt)+" times")
```
