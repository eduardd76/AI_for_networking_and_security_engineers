# Chapter 69: Cloud Security Posture Management
Misconfiguration detection, IAM analysis, cost-risk tradeoffs.


```python
import random
random.seed(42)
print("Cloud Security ready")
```

```python
resources=[{"id":"r-"+str(i),"type":random.choice(["s3","ec2","rds","iam"]),"public":random.choice([True,False,False]),"encrypted":random.choice([True,True,False]),"mfa":random.choice([True,False])} for i in range(20)]
misconfigs=[r for r in resources if r["public"] or not r["encrypted"]]
print("Misconfigurations:",len(misconfigs),"of",len(resources))
for r in misconfigs[:5]: print(" "+r["id"]+" ("+r["type"]+") public="+str(r["public"])+" enc="+str(r["encrypted"]))
```

```python
iam_findings=[{"user":"user-"+str(i),"admin":random.choice([True,False,False,False]),"unused_days":random.randint(0,365),"mfa":random.choice([True,True,False])} for i in range(15)]
risks=[f for f in iam_findings if f["admin"] or not f["mfa"] or f["unused_days"]>90]
print("IAM risks:",len(risks))
print("No MFA:",sum(1 for f in iam_findings if not f["mfa"]))
print("Unused>90d:",sum(1 for f in iam_findings if f["unused_days"]>90))
```
