```python
import random
random.seed(42)
print("AppSec ready")
```

```python
vulns=[{"id":"V-"+str(i),"type":random.choice(["sqli","xss","ssrf","rce","idor"]),"severity":random.choice(["critical","high","medium","low"]),"file":"app/"+random.choice(["auth.py","api.py","db.py","views.py"])} for i in range(25)]
by_sev={}
for v in vulns: by_sev[v["severity"]]=by_sev.get(v["severity"],0)+1
for sev in ["critical","high","medium","low"]: print(sev+":",by_sev.get(sev,0))
```

```python
deps=[{"name":n,"version":v,"cves":c} for n,v,c in [("requests","2.25.0",["CVE-2023-32681"]),("pillow","8.0.0",["CVE-2021-27921","CVE-2021-28675"]),("django","3.2.0",[]),("pyyaml","5.3.0",["CVE-2020-14343"])]]
at_risk=[d for d in deps if d["cves"]]
print("Vulnerable deps:",len(at_risk),"of",len(deps))
for d in at_risk: print(" "+d["name"]+" "+d["version"]+" -> "+str(len(d["cves"]))+" CVEs")
```
