# Chapter 67: Deception Technology and Honeypots
Honeypot placement, attacker profiling, breadcrumb trails.


```python
import random
random.seed(42)
print("Deception Technology ready")
```

```python
honeypots=[{"id":"hp-"+str(i),"type":random.choice(["ssh","http","smb","ftp"]),"ip":"10.0.99."+str(i),"active":True} for i in range(1,6)]
print("Honeypots deployed:",len(honeypots))
for h in honeypots: print(" "+h["id"]+" ("+h["type"]+") at "+h["ip"])
```

```python
interactions=[{"src":"45.33.32."+str(random.randint(1,50)),"honeypot":random.choice([h["id"] for h in honeypots]),"cmd":random.choice(["ls","cat /etc/passwd","wget evil.sh","id","uname -a"])} for _ in range(20)]
attacker_ips=set(i["src"] for i in interactions)
print("Attacker IPs:",len(attacker_ips))
cmd_counts={}
for i in interactions: cmd_counts[i["cmd"]]=cmd_counts.get(i["cmd"],0)+1
print("Top commands:")
for cmd,cnt in sorted(cmd_counts.items(),key=lambda x:-x[1])[:3]: print(" "+cmd+":",cnt)
```
