# Chapter 73: Digital Forensics and Evidence Collection
Memory acquisition, artifact extraction, chain of custody.


```python
import random, hashlib
random.seed(42)
print("Digital Forensics ready")
```

```python
artifacts=[{"name":a,"size":random.randint(1000,10000000),"md5":hashlib.md5(a.encode()).hexdigest()} for a in ["memory.dmp","pagefile.sys","","SAM","SYSTEM","ntuser.dat"]]
print("Artifacts collected:",len(artifacts))
for a in artifacts: print(" "+a["name"]+" md5="+a["md5"][:16]+"...")
```

```python
timeline_events=[{"ts":"2024-03-15T"+str(h).zfill(2)+":00:00","type":t,"file":f} for h,t,f in [(2,"created","dropper.exe"),(2,"exec","dropper.exe"),(3,"created","payload.dll"),(4,"modified","registry"),(5,"deleted","dropper.exe")]]
print("Timeline events:",len(timeline_events))
for e in timeline_events: print(" "+e["ts"]+" "+e["type"]+" "+e["file"])
```
