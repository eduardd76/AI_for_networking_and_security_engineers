```python
random.seed(42)
base=datetime(2024,1,15,10,0,0)
events=[{"time":base+timedelta(seconds=i*random.randint(5,120)),"type":random.choice(["auth_fail","port_scan","file_access"]),"src":"192.168.1."+str(random.randint(1,20)),"severity":random.randint(1,10)} for i in range(50)]
tc={}
for e in events: tc[e["type"]]=tc.get(e["type"],0)+1
print("Event types:")
for t,c in sorted(tc.items()): print(" "+t+":",c)
```
