```python
import random
random.seed(42)
print("AI Threat Detection ready")
```

```python
ATTCK={"T1595":"Active Scanning","T1566":"Phishing","T1059":"Command Scripting","T1053":"Scheduled Task","T1048":"Exfiltration"}
events=[{"type":random.choice(["scan","phish","exec","persist","exfil"]),"src":"10.0.1."+str(random.randint(1,10)),"conf":round(random.uniform(0.4,0.99),2)} for _ in range(50)]
stage_map={"scan":"recon","phish":"access","exec":"execution","persist":"persistence","exfil":"exfiltration"}
seen=set(stage_map[e["type"]] for e in events)
print("Stages:",seen)
print("Chain:",round(len(seen)/5*100),"%")
print("Verdict:","HIGH" if len(seen)>3 else "MEDIUM")
```

```python
def calibrate(raw,base=0.05):
    return round(raw*base/(raw*base+(1-raw)*(1-base)),4)
for s in [0.9,0.75,0.5,0.3,0.1]:
    print("raw="+str(s)+" calibrated="+str(calibrate(s)))
```
