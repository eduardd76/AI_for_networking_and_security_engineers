# Chapter 59: Intrusion Detection with ML
Feature engineering, IDS classifier, precision/recall evaluation.


```python
import numpy as np
import random
print("IDS ML ready")
```

```python
random.seed(42)
n=1000
X=[]; y=[]
for _ in range(n):
    attack=random.random()<0.1
    if attack:
        X.append([random.randint(5000,50000),random.randint(200,5000),round(random.uniform(0.001,0.1),4),random.randint(1,65535)])
    else:
        X.append([random.randint(100,3000),random.randint(1,100),round(random.uniform(0.1,30.0),4),random.choice([80,443,22,53])])
    y.append(1 if attack else 0)
print("Dataset: n="+str(n)+" attack_rate="+str(round(sum(y)/n*100,1))+"%")
```

```python
split=int(n*0.8)
X_train,X_test,y_train,y_test=X[:split],X[split:],y[:split],y[split:]
def predict(x): score=x[0]/50000+x[1]/5000+(1-x[2]/30); return 1 if score>0.3 else 0
preds=[predict(x) for x in X_test]
tp=sum(1 for p,t in zip(preds,y_test) if p==1 and t==1)
fp=sum(1 for p,t in zip(preds,y_test) if p==1 and t==0)
fn=sum(1 for p,t in zip(preds,y_test) if p==0 and t==1)
tn=sum(1 for p,t in zip(preds,y_test) if p==0 and t==0)
prec=tp/(tp+fp) if tp+fp>0 else 0
rec=tp/(tp+fn) if tp+fn>0 else 0
f1=2*prec*rec/(prec+rec+1e-9)
print("Precision:",round(prec,3),"Recall:",round(rec,3),"F1:",round(f1,3))
print("TP:",tp,"FP:",fp,"FN:",fn,"TN:",tn)
```
