# Chapter 62: Automated Threat Hunting
Hypothesis-driven hunting, IOC matching, beaconing detection.


```python
import random
random.seed(42)
print("Threat Hunting ready")
```

```python
IOCS={"ips":{"185.220.101.1","45.33.32.156"},"domains":{"evil.com","malware.xyz"}}
logs=[{"src":"10.0.1."+str(random.randint(1,50)),"dst":random.choice(["8.8.8.8","185.220.101.1","1.1.1.1"]),"domain":random.choice(["google.com","evil.com","cloudflare.com"])} for _ in range(100)]
hits=[l for l in logs if l["dst"] in IOCS["ips"] or l["domain"] in IOCS["domains"]]
print("IOC hits:",len(hits))
for h in hits[:3]: print(" src="+h["src"]+" dst="+h["dst"])
```

```python
src_c={}
for l in logs: src_c[l["src"]]=src_c.get(l["src"],0)+1
beacons=[(s,c) for s,c in src_c.items() if c>3]
print("Beaconing suspects:",len(beacons))
for s,c in sorted(beacons,key=lambda x:-x[1])[:5]: print(" "+s+": "+str(c))
print("Threat hunting complete.")
```
