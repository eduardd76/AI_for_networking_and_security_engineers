Image scanning, runtime threat detection, RBAC analysis.


```python
import random
random.seed(42)
print("Container Security ready")
```

```python
images=[{"name":"myapp:"+v,"vulns":random.randint(0,15),"critical":random.randint(0,3)} for v in ["1.0","1.1","2.0","latest"]]
for img in images:
    risk="BLOCK" if img["critical"]>0 else "WARN" if img["vulns"]>5 else "OK"
    print(img["name"]+": vulns="+str(img["vulns"])+" critical="+str(img["critical"])+" -> "+risk)
```

```python
pods=[{"name":"pod-"+str(i),"privileged":random.choice([True,False,False,False]),"host_net":random.choice([True,False,False]),"root":random.choice([True,False])} for i in range(10)]
risky=[p for p in pods if p["privileged"] or p["host_net"] or p["root"]]
print("Risky pods:",len(risky),"of",len(pods))
for p in risky: print(" "+p["name"]+" privileged="+str(p["privileged"]))
```
