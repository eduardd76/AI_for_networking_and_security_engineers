# Chapter 74: Cyber Threat Intelligence
IOC enrichment, threat actor profiling, intelligence sharing.


```python
import random
random.seed(42)
print("CTI ready")
```

```python
iocs=[{"value":"45.33."+str(random.randint(10,99))+"."+str(random.randint(1,255)),"type":"ip","confidence":round(random.uniform(0.5,0.99),2),"actor":random.choice(["APT28","Lazarus","FIN7","TA505"])} for _ in range(15)]
actors={}
for ioc in iocs: actors[ioc["actor"]]=actors.get(ioc["actor"],0)+1
print("IOCs collected:",len(iocs))
for actor,cnt in sorted(actors.items(),key=lambda x:-x[1]): print(" "+actor+":",cnt,"IOCs")
```

```python
ttps={"APT28":["T1566","T1059","T1027"],"Lazarus":["T1195","T1059","T1048"],"FIN7":["T1566","T1204","T1055"]}
top_actor=max(actors,key=actors.get)
print("Primary threat actor:",top_actor)
if top_actor in ttps:
    print("Known TTPs:",ttps[top_actor])
print("High-conf IOCs:",sum(1 for i in iocs if i["confidence"]>0.8))
```
