# Chapter 64: Behavioral Analytics and UEBA
User baseline modeling, peer group comparison, risk scoring.


```python
import random
random.seed(42)
print("UEBA ready")
```

```python
users=["alice","bob","carol","dave","eve"]
baselines={}
for u in users:
    baselines[u]={"avg_files":random.randint(10,50),"avg_bytes":random.randint(10000,100000),"hours":list(range(8,18))}
    print("Baseline "+u+": avg_files="+str(baselines[u]["avg_files"]))
```

```python
def risk(user,today):
    b=baselines[user]; s=0
    if today["files"]>b["avg_files"]*3: s+=3
    if today["bytes"]>b["avg_bytes"]*5: s+=4
    if today["fails"]>5: s+=3
    if today["hour"] not in b["hours"]: s+=2
    return s
session={"hour":2,"files":500,"bytes":500000,"fails":20}
for u in users:
    r=risk(u,session)
    print(u+": risk="+str(r)+" level="+("HIGH" if r>=7 else "MEDIUM" if r>=4 else "LOW"))
```
