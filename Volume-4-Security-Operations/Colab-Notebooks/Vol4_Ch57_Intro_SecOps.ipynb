SOC metrics, alert analysis, MTTD/MTTR calculation.


```python
import random
print("AI SecOps ready")
```

```python
random.seed(42)
alerts=[random.randint(800,1500) for _ in range(30)]
true_pos=[random.randint(10,50) for _ in range(30)]
fpr=1-sum(true_pos)/sum(alerts)
print("Avg daily alerts:",round(sum(alerts)/len(alerts),1))
print("False positive rate:",round(fpr*100,1),"%")
```

```python
mttd=[random.uniform(0.5,8.0) for _ in range(30)]
mttr=[random.uniform(1.0,24.0) for _ in range(30)]
print("MTTD:",round(sum(mttd)/len(mttd),1),"hours")
print("MTTR:",round(sum(mttr)/len(mttr),1),"hours")
coverage={"endpoint":0.92,"network":0.85,"cloud":0.78,"identity":0.88}
for domain,pct in coverage.items(): print("  "+domain+":",str(round(pct*100))+"%")
```
