```python
import random, json
from datetime import datetime, timedelta
random.seed(42)
print("Incident Response ready")
```

```python
incident={"id":"IR-2024-001","type":"ransomware","severity":"critical","affected_hosts":["srv01","srv02","ws10"]}
playbook=["isolate_hosts","preserve_evidence","notify_stakeholders","eradicate","recover","lessons_learned"]
for i,step in enumerate(playbook,1):
    print(f"Step {i}: {step} - EXECUTING")
print("Playbook complete:",incident["id"])
```

```python
base=datetime(2024,3,15,2,0,0)
timeline=[{"t":base+timedelta(minutes=m),"event":e} for m,e in [(0,"initial_access"),(15,"lateral_movement"),(45,"data_staging"),(90,"exfiltration"),(120,"ransomware_deployed")]]
print("Incident Timeline:")
for t in timeline: print(" "+t["t"].strftime("%H:%M")+" - "+t["event"])
```
