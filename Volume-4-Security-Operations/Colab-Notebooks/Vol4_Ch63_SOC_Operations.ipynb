# Chapter 63: AI-Augmented SOC Operations
Alert triage, priority queuing, workload reduction metrics.


```python
import random
random.seed(42)
print("SOC Operations ready")
```

```python
alerts=[{"id":"A-"+str(i),"type":random.choice(["malware","phishing","anomaly","policy"]),"severity":random.randint(1,10),"conf":round(random.uniform(0.3,0.99),2)} for i in range(100)]
critical=[a for a in alerts if a["severity"]>=8]
print("Total:",len(alerts),"Critical:",len(critical))
```

```python
def triage(a):
    s=a["severity"]*a["conf"]
    if s>=8: return "critical"
    if s>=5: return "high"
    if s>=3: return "medium"
    return "low"
triaged=[{**a,"priority":triage(a)} for a in alerts]
auto_closed=[a for a in triaged if a["severity"]<3 and a["conf"]<0.5]
print("Auto-closed:",len(auto_closed))
print("Workload reduction:",round(len(auto_closed)/len(alerts)*100,1),"%")
```
